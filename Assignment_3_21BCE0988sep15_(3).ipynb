## 5.Outlier Detection

```python
sns.boxplot(df.Fare)
```

INFERENCE: Here We can see there are some outliers, We have to follow the outliers removing techniques and remove them

1.Inter Quatile Range(IQR Method)

2.Z-Score Method

3.Percentile Method

we can remove the outliers by using the any one of those methods.

Outlier removal by  replacement with median

```python
q1=df.Fare.quantile(0.25)
q3=df.Fare.quantile(0.75)
```

```python
q1
```

```python
q3
```

```python
IQR=q3-q1
IQR
```

```python
upper_limit=q3+1.5*IQR
upper_limit
```

```python
fare_outliers=(df["Fare"]>upper_limit).sum()
fare_outliers
```

```python
median_Fare=df.Fare.median()
median_Fare
```

### Outlier Removal

```python
df["Fare"]=np.where(df["Fare"]>upper_limit,median_Fare,df["Fare"])
```

```python
sns.boxplot(df.Fare)
```
